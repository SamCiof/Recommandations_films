### Fusion des deux jeux de données

```python
#Fusion des deux dataframes :

merged = imdb.merge(movielens, on='imdbId', how='outer', suffixes=('_imdb', '_movielens'))
```

### Gestion des colonnes communes

```python
#Gestion des colonnes communes title, genres et averageRating:

#Sachant que la note imbd est reliée à un nombre de vote pour ne pas fausser les données nous 
#garderons la note imdb. La données IMDB étant la plus fiable en terme de recherche de film,
#nous décidons de prioriser la donnée imdb pour les films se trouvant dans les deux datasets.

merged['title'] = merged['title_imdb'].combine_first(merged['title_movielens'])
merged['genres'] = merged['genres_imdb'].combine_first(merged['genres_movielens'])
merged['averageRating'] = merged['averageRating_imdb'].combine_first(merged['averageRating_movielens'])
```

```python
#Suppression des colonnes nous ayant servi à la fusion précédente :
merged.drop(columns=['title_imdb', 'title_movielens', 'genres_imdb', 'genres_movielens', 'averageRating_imdb', 'averageRating_movielens'], inplace=True)
```

```python
#Suppression des 2343 lignes qui n'étaient présentes que dans movielens:
merged.dropna(subset=['numVotes'], inplace=True)
```

### Suppression des lignes comportant des valeurs manquantes

```python
#Test de suppression des lignes dont le genre, le réalisateur, le producteur et les acteurs sont vides, car ces informations vont nous servir pour faire le modèle:

merged_clean = merged.dropna(subset=['genres', 'director', 'actor', 'actress', 'producer'])
merged_clean.shape
```

```python
merged_clean.isna().sum()
```

### Fusion des colonnes actor et actress

```python
# Fusion des deux colonnes 
merged_clean['actor_actress'] = merged_clean['actor'].fillna('') + '/' + merged_clean['actress'].fillna('')
```

```python
#Supression des colonnes actor et actress originales
merged_clean.drop(columns=['actor', 'actress'], inplace=True)
```

```python
merged_clean.head()
```

### Réorganisation du jeu de données

```python
# #Reordonnement des colonnes :
nouvel_ordre = ['imdbId', 'title', 'genres', 'averageRating', 'numVotes', 'director', 'actor_actress', 'producer']

merged_clean = merged_clean[nouvel_ordre]
merged_clean.head()
```

### Transformation de la colonne genres en boléens

```python
#Désormais nous voulons séparer les différents genres de films afin d'y remplir la valeur par des valeurs boléennes / encodage "one-hot :

#Verification des différents genres existant :
import re
pattern = re.compile(r'\s*[,|]\s*|\s*-\s*')

merged_clean['genres_split'] = merged_clean['genres'].apply(lambda x: re.split(pattern, x))

merged_clean['genres_cleaned'] = merged_clean['genres_split'].apply(lambda x: ','.join([genre.strip() for genre in x]))

unique_genres = merged_clean['genres_cleaned'].apply(lambda x: x.split(',')).explode().unique()
unique_genres = sorted(unique_genres) 
unique_genres
```

```python
genres_dummies = merged_clean['genres_cleaned'].str.get_dummies(sep=',')
merged_clean = pd.concat([merged_clean, genres_dummies], axis=1)

merged_clean.head()
```

```python
# Verification de la colonne nan pour vérifier qu'il n'y a pas de valeurs manquantes :
nan_genre_rows = merged_clean[merged_clean['genres'] == 'nan']
nan_genre_rows
```

```python
#Suppression de la colonne nan et de la colonne original genres :
merged_clean.drop(columns=['genres', 'nan', 'genres_split', 'genres_cleaned'], inplace=True)
merged_clean.head()
```
