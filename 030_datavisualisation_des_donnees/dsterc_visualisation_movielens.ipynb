### Importation des librairies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

### Chargement des fichiers

```python
source_dir= '/home/dstrec/dstrec/010_data/001_transformed'

file_movielens = f"{source_dir}/movielens.csv"
```

```python
movielens = pd.read_csv(file_movielens)
```

### Visualisation des données

```python
#Histogramme des 10 genres les plus courants
#############################################

# Séparation des genres et comptage des occurrences
genres = movielens['genres'].str.split(',').tolist()
genres_flat = [genre.strip() for sublist in genres for genre in sublist]
genre_counts = Counter(genres_flat)

# Conversion en DataFrame pour faciliter le tri et le graphique
genre_counts_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False)

# Prendre les 10 genres les plus représentés
top_10_genres = genre_counts_df.head(10)

# Affichage de l'histogramme pour les 10 genres les plus représentés
plt.figure(figsize=(12, 8))
plt.bar(top_10_genres['Genre'], top_10_genres['Count'], color='skyblue')
plt.xlabel('Genres')
plt.ylabel('Nombre de films')
plt.title('Histogramme des 10 genres les plus courants')
plt.xticks(rotation=90)
plt.show()

```

```python
#Histogramme du top 20 des films les mieux notés
################################################

# Trier les films par note moyenne en ordre décroissant et prendre les 20 meilleurs
top_20_movies = movielens.sort_values(by='average_rating', ascending=False).head(20)

# Affichage de l'histogramme pour les 20 films les mieux notés
plt.figure(figsize=(12, 10))
plt.barh(top_20_movies['title'], top_20_movies['average_rating'], color='skyblue')
plt.xlabel('Note moyenne')
plt.ylabel('Titre du film')
plt.title('Top 20 des films les mieux notés')
plt.gca().invert_yaxis()  # Inverser l'axe y pour avoir le film avec la meilleure note en haut
plt.show()
```

```python
#Histogramme de la distribution des notes moyennes
##################################################

# Affichage de l'histogramme de la distribution des notes moyennes
plt.figure(figsize=(12, 8))
sns.histplot(movielens['average_rating'], bins=20, kde=True, color='skyblue')
plt.xlabel('Note moyenne')
plt.ylabel('Fréquence')
plt.title('Distribution des notes moyennes')
plt.show()
```

```python
#Histogramme des 20 tags les plus courants
##########################################

# Séparation des tags et comptage des occurrences
tags = movielens['most_common_tag'].str.split(',').tolist()
tags_flat = [tag.strip() for sublist in tags for tag in sublist]
tag_counts = Counter(tags_flat)

# Conversion en DataFrame pour faciliter le tri et le graphique
tag_counts_df = pd.DataFrame(tag_counts.items(), columns=['Tag', 'Count']).sort_values(by='Count', ascending=False)

# Prendre les 20 tags les plus représentés
top_20_tags = tag_counts_df.head(20)

# Affichage de l'histogramme pour les 20 tags les plus courants
plt.figure(figsize=(12, 8))
plt.bar(top_20_tags['Tag'], top_20_tags['Count'], color='skyblue')
plt.xlabel('Tags')
plt.ylabel('Nombre de films')
plt.title('Histogramme des 20 tags les plus courants')
plt.xticks(rotation=90)
plt.show()
```
